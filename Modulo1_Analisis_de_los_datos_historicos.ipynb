```python
# Función para crear el gráfico
def make_graph(data):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Close'], label='Precio de Cierre', color='blue')
    plt.title('Gráfico de Acciones de Tesla (TSLA)')
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre (USD)')
    plt.legend()
    plt.grid()
    plt.show()

# Descargar datos de acciones de Tesla
tesla_symbol = 'TSLA'
tesla_data = yf.download(tesla_symbol, start='2020-01-01', end='2024-10-25')

# Llamar a la función para graficar
make_graph(tesla_data)
```

```python
# Función para crear el gráfico
def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Close'], label='Precio de Cierre', color='red')
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre (USD)')
    plt.legend()
    plt.grid()
    plt.show()

# Descargar datos de acciones de GameStop
gamestop_symbol = 'GME'
gamestop_data = yf.download(gamestop_symbol, start='2020-01-01', end='2024-10-25')

# Llamar a la función para graficar
make_graph(gamestop_data, 'Gráfico de Acciones de GameStop (GME)')
```
